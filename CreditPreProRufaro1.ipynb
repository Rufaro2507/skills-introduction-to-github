```python
#term and employment length expected are text strings, need to convert them to numeric using str.replace method
loan_data['emp_length'].unique()
```

```python
#store the new emp length in a new variable called loan_data['emp_length_int']
loan_data['emp_length_int'] = loan_data['emp_length'].str.replace('10+ years', '10')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('< 1 year', '0')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace('n/a', '0')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' years', '')
loan_data['emp_length_int'] = loan_data['emp_length_int'].str.replace(' year', '')
```

```python
#The expression type(loan_data['emp_length_int'][0]) will return the type of the first element in the emp_length_int column of the loan_data DataFrame.
type(loan_data['emp_length_int'][0])
```

```python
#to transfer emp lenght int to a numeric value we use the pands function pd to numeric
#loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'])
loan_data['emp_length_int'] = pd.to_numeric(loan_data['emp_length_int'])
```
